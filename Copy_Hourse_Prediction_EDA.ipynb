## Data Cleaning

```python
# Collecting columns with missing data into a list

# Initialize variables
missing_columns = []
count = 0  # Counter for columns with any missing values

# Iterate through each column in the dataframe
for feature in df.columns:
    missing_percentage = (df[feature].isnull().sum() / len(df)) * 100

    if missing_percentage > 0:
        count += 1
        missing_columns.append(feature)
        print(f"Missing percentage of {feature}: {missing_percentage:.4f}%")

# Output the results
print(f"Total columns with missing data: {count}")
```

```python
# List of columns with missing data


# Calculate missing and available data percentages
missing_percentage = df[missing_columns].isnull().mean() * 100
available_percentage = 100 - missing_percentage

# Plot
fig, ax = plt.subplots(figsize=(12, 4))

# Plot each column with missing and available data percentages
bar_width = 0.6
bar_positions = np.arange(len(missing_columns))

# Plot stacked bars
bars1 = ax.bar(bar_positions, available_percentage, bar_width, color='green', label='Available Data')
bars2 = ax.bar(bar_positions, missing_percentage, bar_width, bottom=available_percentage, color='red', label='Missing Data ')

# Customize the plot
ax.set_xticks(bar_positions)
ax.set_xticklabels(missing_columns, rotation=45, ha='right')
ax.set_title('Missing Data and Available Data')
ax.set_ylabel('Percentage')
ax.set_xlabel('Column')
ax.legend(title='Data Type')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# List to store columns with more than 40% missing values

missing_percentage_40 = [feature for feature in missing_columns if ((df[feature].isnull().sum()) / (len(df)) * 100) > 40]  # List to store columns with more than 40% missing values

print(f"Columns with more than 40% missing data: {missing_percentage_40}")
```

```python
# Dropping columns with more than 40% missing data + Id column

missing_percentage_40 = missing_percentage_40 + ["Id"] # droping the "Id" column
new_df = df.drop(missing_percentage_40, axis = 1)
new_df.shape
```

```python
# Separating columns into categorical, numerical, and numerical-categorical
# Applying mode to categorical columns and median to numerical columns for missing data imputation
# Median is used instead of mean for numerical columns to reduce the impact of outliers


columns = new_df.columns

categorical_columns = []
numerical_columns = []
numerical_categorical_columns = []


# Loop through each feature in the dataset
for feature in columns:
    if new_df[feature].dtype == "O":
        # Categorical column (non-numerical)
        categorical_columns.append(feature)
    elif new_df[feature].nunique() < 30:
        # Numerical but with a small number of unique values, treated as categorical
        numerical_categorical_columns.append(feature)
    else:
        # Numerical column with many unique values
        numerical_columns.append(feature)



# List of substrings to check in column names
year = ["Yr", "Year", "Mo"]

# Filtering numerical columns to find those that contain any of the substrings in the 'year' list
year_columns = [feature for feature in numerical_columns + numerical_categorical_columns if any(x in feature for x in year)]

# Remove elements in year_columns from numerical_columns
numerical_columns = [feature for feature in numerical_columns if feature not in year_columns]

# Remove elements in year_columns from numerical_categorical_columns
numerical_categorical_columns = [feature for feature in numerical_categorical_columns if feature not in year_columns]


# Display the number of unique values in each category
print("Categorical Columns:")
for feature in categorical_columns:
    print(f"{feature} --> {new_df[feature].nunique()} unique values")

print("\nNumerical Categorical Columns:")
for feature in numerical_categorical_columns:
    print(f"{feature} --> {new_df[feature].nunique()} unique values")

print("\nNumerical Columns:")
for feature in numerical_columns:
    print(f"{feature} --> {new_df[feature].nunique()} unique values")
    
print("\nyear Columns:")
for feature in year_columns:
    print(f"{feature} --> {new_df[feature].nunique()} unique values")
```
